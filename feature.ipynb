##  Getting started with extreacting feature

```python


def extract_episode(txt):
    check = False
    data = ""
    for i in txt:
        if i == '(':
            check = True
        if check == True:
            data += i
        if i == ')':
            break
    return data
```

```python
df["Episodes"]=df["Title"].apply(extract_episode)
```

```python


# 2. Clean the string: remove '(', ')', and 'eps', then strip whitespace
df['Episodes'] = df['Episodes'].str.replace('(', '').str.replace(')', '').str.replace('eps', '').str.strip()

```

```python
df['Episodes']=  df['Episodes'].astype(int)
```

```python
def extraction_time(txt):
    if not isinstance(txt, str):
        return None

    pos = txt.find(')')
    if pos == -1:
        return None

    return txt[pos+1 : pos+18]
```

```python
df['Total Time'] = df['Title'].apply(extraction_time)
```
